```python
#import Libraries
import json
import pandas
import re
```

```python
#read file
file = r'1.txt'
f = open(file)
k = f.read()
print(k)
```

```python
# read each item from i
i = re.findall(r"URL.*?\n\n",k,flags=re.S)
for item in i:
    print(item)
```

```python
#create a list to append each value

url,mention,mention_value,mention_id,mention_url,token,token_value,token_key = [],[],[],[],[],[],[],[]
for item in i:
    url.append(re.findall(r"URL\t(.*)",item))
    mention.append(re.findall(r"MENTION.*",item))
    mention_value.append(re.findall(r"MENTION\t(.*?)\t",item))
    mention_id.append(re.findall(r"MENTION\t.*?\t(.*?)\t",item))
    mention_url.append(re.findall(r"MENTION.*\t(.*)",item))
    token.append(re.findall(r"TOKEN.*",item))
    token_value.append(re.findall(r"TOKEN\t(.*)\t",item))
    token_key.append(re.findall(r"TOKEN.*\t(.*)",item))

print(mention_url)
    
```

```python
#create a funtion to append value  inside mention and token 

def json_create(url,mention_value,mention_id,mention_url,token_value,token_key):
    sample1 = {"URL":"","MENTION":[],"TOKEN":[]}

    # for i in range(len(mention_value)):
    #       sample1["MENTION"].append({"value":"","id":"","url":""})

    sample1["URL"] = url[0]
    

    [sample1["MENTION"].append({"value":"","id":"","url":""}) for i in range(len(mention_value))]
    for i in range(len(mention_value)):
        sample1["MENTION"][i]["value"] = mention_value[i]
        sample1["MENTION"][i]["id"] = mention_id[i]
        sample1["MENTION"][i]["url"] = mention_url[i]

    [sample1["TOKEN"].append({"value":"","id":""}) for i in range(len(token_value))]

    for i in range(len(token_value)):
        sample1["TOKEN"][i]["value"] = token_value[i]
        sample1["TOKEN"][i]["id"] = token_key[i]
    return sample1
```

```python
# url,mention,mention_value,mention_id,mention_url,token,token_value,token_key = [],[],[],[],[],[],[],[]

out = json_create(url[0],mention_value[0],mention_id[0],mention_url[0],token_value[0],token_key[0])
```

```python
# printing  values of first index
out
```

```python
# looping and appending output to final_out
final_out = []
for i in range(len(url)):
    out = json_create(url[i],mention_value[i],mention_id[i],mention_url[i],token_value[i],token_key[i])
    final_out.append(out)
```

```python
#saving output as json file

f = open('output.json','w')
json.dump(final_out, f)
f.close()
```
